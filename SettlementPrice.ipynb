```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(".jupyter\TWN2_stocks.csv")
df.head(60)
k = 1
```

```python
df["PriceNextMonth"] = df["Adj Close"].shift(-k)
```

```python
df.tail(60)
```

```python
dates = np.array(df["Date"])
dates_check = dates[-k:]
dates = dates[:-k]
print(dates_check)
print(dates)
```

```python
X = np.array(df.drop(["PriceNextMonth", "Date"], 1))
X_Check = X[-k:]
X = X[:-k]
df.dropna(inplace = True)
y = np.array(df["PriceNextMonth"])
print(X_Check)
print(X)
print(y)
```

```python
from sklearn.model_selection import cross_validate
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = cross_validation.train_test_split(X,y,test_size = 0.2)
```

```python
model = RandomForestRegressor(criterion='mse',n_jobs=-1,n_estimators = 100)
model.fit(X_train, y_train)
```

```python
conf = model.score(X_test, y_test)
print(conf)
```
